```python
df.drop(columns=['Id', 'MSSubClass','OverallCond','BsmtFinSF2','LowQualFinSF', 'BsmtHalfBath','KitchenAbvGr','EnclosedPorch','MiscVal','YrSold'], inplace=True)
```

```python
new_numerical_features = df.select_dtypes(include=['int64', 'float64']).columns
df[new_numerical_features].corr()['SalePrice'].sort_values(ascending=False)
```

```python
scaled_data = scale.fit_transform(df[new_numerical_features].dropna())
scaled_data
```

```python
scaled_df = pd.DataFrame(scaled_data, columns=new_numerical_features)
scaled_df.head()
```

```python
msk = np.random.rand(len(scaled_df)) <= 0.8

train = scaled_df[msk]
test = scaled_df[~msk]
```

```python
X = train.drop(columns=['SalePrice']).dropna()
y = train['SalePrice'].loc[X.index].dropna()   # keep alignment

X_test = test.drop(columns=['SalePrice']).dropna()
y_test = test['SalePrice'].loc[X_test.index].dropna()

model = linear_model.LinearRegression()
model.fit(X, y)

z = model.predict(X_test)

print(z)


mse = metrics.mean_squared_error(y_test, z)
r = metrics.r2_score(y_test, z)
r_sqrt = np.sqrt(mse)

print("Mean Squared Error:", mse)
print("R^2 Score:", r)
print("Root Mean Squared Error:", r_sqrt)
```

```python
lesso = linear_model.Lasso(alpha=0.1)

lesso.fit(X, y)
z_lasso = lesso.predict(X_test)



mse = metrics.mean_squared_error(y_test, z_lasso)
r = metrics.r2_score(y_test, z_lasso)
r_sqrt = np.sqrt(mse)


print(z_lasso)

print("Mean Squared Error:", mse)
print("R^2 Score:", r)
print("Root Mean Squared Error:", r_sqrt)
```

```python
ridge = linear_model.Ridge(alpha=0.1)

ridge.fit(X, y)
z_ridge = ridge.predict(X_test)



mse = metrics.mean_squared_error(y_test, z_ridge)
r = metrics.r2_score(y_test, z_ridge)
r_sqrt = np.sqrt(mse)


print(z_ridge)

print("Mean Squared Error:", mse)
print("R^2 Score:", r)
print("Root Mean Squared Error:", r_sqrt)
```

```python
polynomial = PolynomialFeatures(degree=2)
x_poly = polynomial.fit_transform(X)
x_poly_test = polynomial.transform(X_test)


model = linear_model.LinearRegression()
model.fit(x_poly, y)

y_poly = model.predict(x_poly)
y_poly_test = model.predict(x_poly_test)
mse = metrics.mean_squared_error(y_test, y_poly_test)
r_squared_test = metrics.r2_score(y_test, y_poly_test)
r_squared_poly = metrics.r2_score(y, y_poly)


r_sqrt = np.sqrt(mse)

print(y_poly)
print(y_poly_test)
print("Mean Squared Error:", mse)
print("R^2 Score on Test Data:", r_squared_test)
print("R^2 Score on Training Data:", r_squared_poly)
print("Root Mean Squared Error:", r_sqrt)
```
